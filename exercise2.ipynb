## 9. Average Tesla Stock Closing Price in January 2025
1. Using the Tesla Stock Data 2025 dataset, analyze the closing prices for the latest available month (January 2025) and visualize the results.
2. Filter the dataset to include only records from January 2025.
3. Plot the closing prices using either a bar chart or a box plot to display the distribution of stock prices in January 2025.

  Ensure you use pandas for data manipulation and matplotlib/seaborn for visualization.
    
     

```python
# 1. Load data (creating sample data for demonstration)
tesla_data = pd.DataFrame({
    'Date': pd.date_range(start='2025-01-01', end='2025-01-31'),
    'Close': [220, 225, 218, 230, 235, 240, 238, 245, 242, 250,
              248, 255, 253, 260, 258, 265, 263, 270, 268, 275,
              273, 280, 278, 285, 283, 290, 288, 295, 293, 300, 305]
})

# 2. Filter data for January 2025
january_data = tesla_data[tesla_data['Date'].dt.month == 1]

# 3. Create box plot of closing prices
plt.figure(figsize=(10, 6))
sns.boxplot(data=january_data, y='Close', color='lightblue')

# Chart title and labels
plt.title('Tesla Stock Closing Prices Distribution - January 2025', fontsize=14)
plt.ylabel('Closing Price ($)', fontsize=12)
plt.xlabel('January 2025', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 4. Calculate and display average closing price
average_price = january_data['Close'].mean()
plt.axhline(y=average_price, color='red', linestyle='--',
            label=f'Average: ${average_price:.2f}')
plt.legend()

plt.show()

# Additional line chart for daily price movement
plt.figure(figsize=(12, 6))
plt.plot(january_data['Date'], january_data['Close'],
         marker='o', color='green', linestyle='-')

plt.title('Daily Tesla Closing Prices - January 2025', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price ($)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 14. Analyzing Blood Pressure Distribution Across Age Groups
Using the Heart Disease Risk Prediction Dataset, visualize the distribution of resting blood pressure across different age groups using a violin plot.

1. Filter the dataset to include only relevant columns: age and resting blood pressure (trestbps).
2. Categorize patients into age groups (e.g., 20-30, 30-40, etc.).
3. Create a violin plot to visualize the distribution of resting blood pressure for each age group.
4. Use the hue parameter to differentiate the distribution based on gender (sex column: male/female).
5. Label the axes and add a title for better clarity.

```python
# 1. Filter the dataset
blood_pressure_data = heart_data[['age', 'trestbps', 'sex']]

# 2. Categorize patients into age groups
blood_pressure_data['age_group'] = pd.cut(blood_pressure_data['age'],
                                         bins=[20, 30, 40, 50, 60, 70, 80],
                                         labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70-79'])

# 3. Create the violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='age_group', y='trestbps', hue='sex', data=blood_pressure_data, split=True, palette='Set1')

# 4. Customize the plot
plt.title('Distribution of Resting Blood Pressure Across Age Groups', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Resting Blood Pressure (mmHg)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Gender')

# 5. Display the plot
plt.tight_layout()
plt.show()
```
